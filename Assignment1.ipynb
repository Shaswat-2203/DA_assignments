```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path="/content/drive/MyDrive/ToyotaCorolla.csv"
```

```python
toyota_data = pd.read_csv(path)
toyota_data.describe()
```

```python
toyota_data.head()
```

```python
toyota_data.info()
```

```python
sns.scatterplot(x='Price',y='Age',data=toyota_data)
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(8,6))
sns.regplot(x = toyota_data["Age"], y = toyota_data["Price"])
plt.show()
```

```python
sns.lmplot(x='Age', y='Price', data=toyota_data,
          fit_reg=False,
          hue='FuelType',
          legend=True,
          palette="Set1",height=6)
```

```python
plt.figure(figsize=(8,6))
sns.distplot(toyota_data['Age'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.distplot(toyota_data['Age'],kde=False)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.distplot(toyota_data['Age'],kde=False,bins=5)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="FuelType", data=toyota_data)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="FuelType", data=toyota_data,
             hue="Automatic")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=toyota_data["Price"])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=toyota_data["FuelType"], 
           y=toyota_data["Price"],
           )
plt.show()  
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="FuelType", 
            y="Price",
            data=toyota_data,
            hue="Automatic"
           )
plt.show()
```
